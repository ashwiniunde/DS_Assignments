```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=3)
```

```python
y_predicted=km.fit_predict(df[['Total Refunds','Total Revenue']])
y_predicted
df['cluster']=y_predicted
```

```python
df1= df[df.cluster==0]
df2= df[df.cluster==1]
df3= df[df.cluster==2]
```

```python
plt.scatter(km.cluster_centers_[:,1],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('Total Refunds')
plt.ylabel('Total Revenue')
plt.legend
```
